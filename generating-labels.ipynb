# Generating Labels Using K-Means

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

% matplotlib inline

from sklearn.cluster import KMeans, MiniBatchKMeans
```

```python
reviews = pd.read_csv('dataset/review.csv', usecols=['text', 'useful', 'cool'])
```

```python
reviews
```

```python
kmeans = KMeans(n_clusters=3)
clusters = kmeans.fit(reviews['useful'].values.reshape(-1,1))
```

```python
clusters.labels_ 
```

```python
clusters.cluster_centers_
```

```python
plt.figure(figsize=(14, 7))

colourmap = np.array(['red', 'green', 'blue'])

plt.scatter(reviews.useful, reviews.cool, c=colourmap[clusters.labels_], s = 40)
```
